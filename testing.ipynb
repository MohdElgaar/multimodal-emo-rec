```python
from facenet_pytorch import MTCNN, InceptionResnetV1
import torch
import numpy as np
import os
import cv2
from glob import glob
import matplotlib.pyplot as plt
# device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
device = torch.device('cpu')
```

```python
data = glob('/usr/cs/public/mohd/2020-1/train/*.mp4')
```

```python
path = data[0]
vidcap = cv2.VideoCapture(path)
success,image = vidcap.read()
count = 0
imgs = []
while success:
  imgs.append(image)
  success,image = vidcap.read()
  count += 1
vidcap.release()
```

```python
mtcnn = MTCNN(select_largest=True).to(device)
resnet = InceptionResnetV1(pretrained='vggface2').eval().to(device)
```

```python
cropped = []
for img in imgs:
    x_aligned, prob = mtcnn(img, return_prob=True)
    if x_aligned is not None:
        cropped.append(x_aligned)
        print('Face detected with probability: {:8f}'.format(prob))
```

```python
from glob import glob
import os
from PIL import Image
import cv2
from tqdm import tqdm
import numpy as np
from PIL import Image
from timeit import timeit
from time import time
```

```python
segment_size = 16
step_size = 10
rootdir = '/usr/cs/public/mohd/data/train/'
files = glob(rootdir + '*_imgs/')
```

```python
def load_batch(ids):
    batch = []
    for idx in ids:
        sample = []
        fn = files[idx]
        frames = os.listdir(fn)
        num_frames = len(frames)
        start_idx = np.random.randint(0, max(num_frames, num_frames - segment_size*step_size))
        for i in range(start_idx, min(num_frames, start_idx + segment_size*step_size), step_size):
            img = Image.open(os.path.join(fn, frames[i]))
            sample.append(np.asarray(img))
        batch.append(sample)
    return batch
```

```python
t0 = time()
ids = np.random.randint(0, len(files), 128)
batch = load_batch(ids)
print(time() - t0)
```
